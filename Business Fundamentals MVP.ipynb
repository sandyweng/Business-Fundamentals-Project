Analysis of Citi Bike Demand

Some initial findings from exploring the Citi Bike dataset were that there were changes in ridership behavior and changes in demand for stations. For example, some of the most popular stations in 2019 decreased in demand in 2020. Also, although the ridership trends throughout the day followed the same trajectory there has been more rides in the afternoon in 2020 compared to 2019. We can conclude from these findings that there have been a change in station demands and Citi Bike should adjust their rebalancing operations to prevent potential discontent and improve user experience.
